# 04. ポートフォリオ分析デモ

```python
from stockkit.analysis import portfolio
from stockkit.viz import charts

symbols = ['AAPL','MSFT','NVDA','GOOGL','7203','9984','6758']
panel = portfolio.price_panel(symbols, period='3y')
portfolio.summary(panel).round(3)
```

```python
charts.cumulative_returns_chart(portfolio.cumulative_returns(panel))
```

```python
charts.correlation_heatmap(portfolio.correlation(panel))
```

```python
# 等加重ポートフォリオ
ew = portfolio.weighted_portfolio(panel)
ew.tail()
```
