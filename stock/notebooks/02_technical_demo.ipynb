# 02. テクニカル分析デモ

SMA/EMA/RSI/MACD/Bollinger/ATR + チャート

```python
from stockkit.data import get_prices
from stockkit.analysis import technical
from stockkit.viz import charts

df = get_prices('NVDA', period='2y')
ind = technical.add_indicators(df)
ind.tail()
```

```python
charts.full_dashboard(df, 'NVDA')
```

```python
# ゴールデン/デッドクロス検出
sig = technical.signal_golden_cross(df)
sig[sig != 0].tail(10)
```
