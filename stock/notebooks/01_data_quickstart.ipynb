# 01. データ取得クイックスタート

- `stockkit.data.get_prices` で OHLCV 取得（DuckDB ローカルキャッシュ自動）
- 4桁数字の銘柄コードは自動で `.T` 付きに正規化

```python
from stockkit.data import get_prices, normalize_symbol

print(normalize_symbol('7203'), normalize_symbol('aapl'))
df = get_prices('AAPL', period='1y')
df.tail()
```

```python
# 日本株
jp = get_prices('7203', period='6mo')
jp.tail()
```

```python
# 2回目はキャッシュから（高速）
import time
t = time.time(); df2 = get_prices('AAPL', period='1y'); print(f'cached: {time.time()-t:.3f}s, rows={len(df2)}')
```
