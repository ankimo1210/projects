# 05. スクリーナーデモ

```python
from stockkit.analysis import screener

universe = ['AAPL','MSFT','NVDA','GOOGL','META','AMZN','TSLA',
            '7203','9984','6758','8306','9432']
rules = [
    screener.pe_below(40),
    screener.above_sma(200),
    screener.rsi_between(35, 75),
]
screener.screen(universe, rules)
```
