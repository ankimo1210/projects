## 価格パネル(前埋めなし整列)

```python
from irp.data import get_prices, price_panel
syms = ['SPY', 'QQQ', 'IWM', 'GLD']
results = {}
for s in syms:
    try:
        results[s] = get_prices(s, '2019-01-01', '2023-12-31')
    except Exception as e:
        print(f'{s}: offline ({type(e).__name__})')
if results:
    panel = price_panel(results)  # union dates, NO ffill
    print('panel shape:', panel.shape)
    print('first valid per symbol:')
    print(panel.apply(lambda c: c.first_valid_index()))
    display(panel.tail(3))
else:
    print('no data downloaded (offline) — universe still defined above')
```
