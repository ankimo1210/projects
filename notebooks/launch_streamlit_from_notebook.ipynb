# 不動産投資シミュレーター — Streamlit ランチャー

このノートブックから Streamlit アプリの起動・停止ができます。

- アプリ本体: `real_estate_app/app.py`
- URL: http://127.0.0.1:8501

## セットアップ（初回のみ）

```python
# 必要パッケージのインストール（初回のみ実行）
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "streamlit", "pandas", "numpy", "matplotlib"])
print("インストール完了")
```

## アプリ起動

```python
import subprocess, sys, os, time

APP_DIR = os.path.join(os.path.dirname(os.path.abspath("__file__")), "real_estate_app")
# Fallback: if running from the notebooks directory
if not os.path.isdir(APP_DIR):
    APP_DIR = os.path.expanduser("~/projects/notebooks/real_estate_app")

proc = subprocess.Popen(
    [sys.executable, "-m", "streamlit", "run", "app.py",
     "--server.address=127.0.0.1", "--server.port=8501",
     "--server.headless=true", "--browser.gatherUsageStats=false"],
    cwd=APP_DIR,
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
)

time.sleep(2)
if proc.poll() is None:
    print(f"Streamlit 起動成功 (PID: {proc.pid})")
    print(f"URL: http://127.0.0.1:8501")
else:
    print("起動失敗")
    print(proc.stderr.read().decode())
```

## アプリ停止

```python
# Streamlit サーバーを停止
try:
    proc.terminate()
    proc.wait(timeout=5)
    print(f"Streamlit 停止完了 (PID: {proc.pid})")
except NameError:
    print("起動済みプロセスが見つかりません。先に起動セルを実行してください。")
except Exception as e:
    print(f"停止エラー: {e}")
    try:
        proc.kill()
        print("強制終了しました")
    except Exception:
        pass
```

## ステータス確認

```python
try:
    if proc.poll() is None:
        print(f"Streamlit 実行中 (PID: {proc.pid})")
        print(f"URL: http://127.0.0.1:8501")
    else:
        print(f"Streamlit 停止済み (return code: {proc.returncode})")
except NameError:
    print("プロセスが開始されていません")
```
