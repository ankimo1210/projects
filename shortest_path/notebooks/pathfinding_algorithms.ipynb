# Shortest Path Algorithms

This notebook compares Dijkstra, A*, bidirectional Dijkstra, and bidirectional A* on the same weighted grid. The Python algorithms emit trace data, and the HTML widget below animates those traces.

```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
project_root = next(
    path for path in (cwd, *cwd.parents)
    if (path / "src" / "shortest_path_lab").exists()
)
src_path = project_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from IPython.display import HTML

from shortest_path_lab.algorithms import (
    a_star,
    bidirectional_a_star,
    bidirectional_dijkstra,
    dijkstra,
)
from shortest_path_lab.examples import make_demo_problem, solve_demo_problem
from shortest_path_lab.grid import GridProblem
from shortest_path_lab.visualization import render_notebook_html
```

## Run the demo grid

The grid has blocked cells and weighted cells. Moving into a weighted cell costs more than moving into a normal cell.

```python
problem = make_demo_problem()
results = solve_demo_problem(problem)

print(f"{'algorithm':28} {'cost':>6} {'expanded':>9} {'path nodes':>10} {'steps':>7}")
print("-" * 66)
for result in results.values():
    expanded = len(result.expanded_forward) + len(result.expanded_backward)
    print(
        f"{result.algorithm:28} {result.cost:6.1f} {expanded:9d} "
        f"{len(result.path):10d} {len(result.steps):7d}"
    )
```

```python
HTML(render_notebook_html(problem, results, height=720))
```

## Try a smaller custom grid

The direct middle row has high costs, so the optimal path should detour around it.

```python
small = GridProblem(
    width=7,
    height=5,
    start=(0, 2),
    goal=(6, 2),
    blocked=frozenset({(3, 0), (3, 1), (3, 3), (3, 4)}),
    weights={(1, 2): 8.0, (2, 2): 8.0, (4, 2): 8.0, (5, 2): 8.0},
)
small_results = {
    "dijkstra": dijkstra(small.neighbors, small.start, small.goal),
    "a_star": a_star(small.neighbors, small.start, small.goal, small.heuristic),
    "bidirectional_dijkstra": bidirectional_dijkstra(
        small.neighbors,
        small.start,
        small.goal,
        reverse_neighbors=small.reverse_neighbors,
    ),
    "bidirectional_a_star": bidirectional_a_star(
        small.neighbors,
        small.start,
        small.goal,
        small.heuristic,
        reverse_neighbors=small.reverse_neighbors,
    ),
}
HTML(render_notebook_html(small, small_results, height=620))
```
