## tl;dr

The registry contains **20 reviewed source candidates**. Only explicitly approved public URLs are enabled for automatic fetch.

## Context & Methods

Grain: one row per source registry entry. The notebook reads `config/sources.yaml`; it does not fetch the web.

```python
from pathlib import Path
from collections import Counter
import yaml
root = Path.cwd()
sources = yaml.safe_load((root/'config/sources.yaml').read_text())['sources']
print('sources:', len(sources))
for field in ['language','source_type','access_type','crawl_policy','copyright_risk']:
    print(field, dict(Counter(row[field] for row in sources)))
```

## Data

The registry stores URL, publisher, access status, crawl policy, rights risk, review date, and notes.

```python
enabled = [row for row in sources if row['enabled']]
[(row['source_id'], row['crawl_policy']) for row in enabled]
```

## Results

Paid, registered, member-only, and uncertain-provenance sources remain disabled.

## Takeaways

New sources should begin disabled and require explicit rights/access review before fetching.
