## Context & Methods

Grain: one extracted question-like text span per source position. Counts include learning-outcome-like statements and extraction fragments.

```python
from pathlib import Path
from collections import Counter
import json
root = Path.cwd()
rows = [json.loads(line) for line in (root/'data/reviewed/questions.jsonl').read_text().splitlines() if line]
print('rows:', len(rows), 'unique ids:', len({row['question_id'] for row in rows}))
for field in ['source_id','language','question_format','topic_primary']:
    print(field, dict(Counter(row.get(field) for row in rows)))
```

## Data

Raw and normalized text stay in ignored local files. This notebook displays only aggregate counts.

```python
print('mark allocation present:', sum(row.get('mark_allocation') is not None for row in rows))
print('low extraction confidence:', sum(row['extraction_confidence'] < .75 for row in rows))
```
